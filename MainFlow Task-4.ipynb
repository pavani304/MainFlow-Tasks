# Exploratory Data Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as datetime
import seaborn as sns
import numpy as np
data = pd.read_csv("/home/pavani/USvideos.csv")
df = pd.DataFrame(data)
df
```

```python
df["trending_date"] = df["trending_date"].apply(lambda x : datetime.datetime.strptime(x, '%y.%d.%m'))
df.head(3)
```

```python
df['publish_time'] = pd.to_datetime(df["publish_time"])
df.head(2)
```

```python
df['publish_month'] = df['publish_time'].dt.month
df['publish_day'] = df['publish_time'].dt.day
df['publish_hour'] = df['publish_time'].dt.hour
df.head(2)
```

```python
df['year'] = df['publish_time'].dt.year
yearly_counts = df.groupby('year')['video_id'].count()
yearly_counts.plot(kind='bar',xlabel='Year',ylabel = 'Total Publish Count',title='Total Publish Video Per Year')
plt.show()
```

# Distribution of Variables

```python
plt.figure(figsize=(12,6))
sns.countplot(x='category_id',data=df,order=df['category_id'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Video Count by Category")
plt.show()
```

```python
views = data['views']

# Plotting the distribution using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(views,color='blue')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
likes = data['likes']

# Creating a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=likes)
plt.title('Boxplot of Likes')
plt.xlabel('Likes')
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Density plot
sns.kdeplot(views, ax=axes[0], color='darkblue', fill=True)
axes[0].set_title('Density Plot of Views')
axes[0].set_xlabel('Views')
axes[0].set_ylabel('Density')

# Violin plot
sns.violinplot(x=views, ax=axes[1], color='lightgreen')
axes[1].set_title('Violin Plot of Views')
axes[1].set_xlabel('Views')

# Show the plots
plt.tight_layout()
plt.show()
```

# Correlation between Variables

## Heatmap

```python
numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']

# Calculating the correlation matrix
correlation_matrix = data[numerical_cols].corr()

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Between Variables')
plt.show()
```

## Bubble Plot

```python
numerical_cols = ['likes', 'dislikes', 'views', 'comment_count']

# Create a scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x='likes', y='dislikes', size='views', hue='comment_count', sizes=(20, 2000), palette='viridis', alpha=0.7)
plt.title('Bubble Plot of Likes vs Dislikes')
plt.xlabel('Likes')
plt.ylabel('Dislikes')
plt.legend(title='Comment Count')
plt.show()
```

## Scatter Plot

```python
numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']

# Create a scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x='likes', y='views', hue='comment_count', size='dislikes', sizes=(20, 200), palette='viridis', alpha=0.7)
plt.title('Scatter Plot of Likes vs Views')
plt.xlabel('Likes')
plt.ylabel('Views')
plt.legend(title='Comment Count')
plt.show()
```

 # Identifying Outliers

```python
numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']

# Create box plots for each numerical column
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=data[col], color='skyblue')
    plt.title(col)
plt.tight_layout()
plt.show()
```
